# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans


        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

## Taking Graph Input from User

```python
def get_user_graph():
    n = int(input("Enter the number of cities: "))
    graph = []
    print(f"Enter the {n}x{n} adjacency matrix (distances). Use 'inf' for no direct path. Press Enter after each row:")
    for i in range(n):
        row_str = input(f"Row {i+1} (space-separated values): ")
        row_values = []
        for val_str in row_str.split():
            if val_str.lower() == 'inf':
                row_values.append(float('inf'))
            else:
                row_values.append(int(val_str))
        graph.append(row_values)

    if len(graph) != n or any(len(row) != n for row in graph):
        print("Error: The input graph is not a valid N x N matrix.")
        return None


user_graph = get_user_graph()

if user_graph:
    min_cost_user, path_user = tsp(user_graph)
    print("\nMinimum travelling cost (user input graph):", min_cost_user)
    print("Optimal path (user input graph):", path_user)
else:
    print("Failed to get a valid graph from user input.")
```
